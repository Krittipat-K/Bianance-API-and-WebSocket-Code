```python
from dotenv import load_dotenv
import os
load_dotenv()
api_key = os.getenv('Binance_API_Test_1_KEY')
api_secret = os.getenv('Binance_API_Test_1_SECRET')
```

Accessing Binance API with Python Library \
Binance API Document : https://binance-docs.github.io/apidocs/futures/en/#change-log

```python
from binance.um_futures import UMFutures
futures_client = UMFutures(key=api_key,secret=api_secret) 
```

Checking Future Account \
Binance API Document : https://binance-docs.github.io/apidocs/futures/en/#account-information-v2-user_data

```python
futures_client.account()
```

Checking Future Account Balance \
Binance API Document : https://binance-docs.github.io/apidocs/futures/en/#futures-account-balance-v2-user_data

```python
for i in futures_client.balance():
    
    if i['asset'] == 'USDT':
        
        print(i)
```

Place New Order \
Binance API Document : https://binance-docs.github.io/apidocs/futures/en/#new-order-trade

```python
params = {
    'symbol': 'BTCUSDT',
    'side': 'BUY',
    'type': 'LIMIT',
    'positionSide':'LONG',
    'timeInForce': 'GTC',
    'quantity': 0.003,
    'price': 39950.90
}

response = futures_client.new_order(**params)
print(response)
```

Set Limit Price \
Binance API Document : https://binance-docs.github.io/apidocs/futures/en/#new-order-trade

```python
params = {
    'symbol': 'BTCUSDT',
    'side': 'SELL',
    'type': 'LIMIT',
    'positionSide':'LONG',
    'timeInForce': 'GTC',
    'quantity': 0.003, # minimum size
    'price': 40000
}

response = futures_client.new_order(**params)
print(response)
```
