Buy Spot \
Binance API Document : https://binance-docs.github.io/apidocs/spot/en/#new-order-trade \
Binance API Python Document : https://binance-connector.readthedocs.io/en/latest/binance.spot.trade.html#new-order-trade

```python
# Post a new order
params = {
    'symbol': 'BTCUSDT',
    'side': 'BUY',
    'type': 'LIMIT',
    'timeInForce': 'GTC',
    'quantity': 0.00013, # notion minimum is 5 USDT so price / quantity >= 5
    'price': 42723.05
}
response = client.new_order(**params)
print(response)
```
